```python
import pandas as pd
import wikipedia
```

```python
articles=['Linear Algebra', 'Data Science','Artificial intelligence','European Central Bank','Financial technology','International Monetary Fund','Basketball','Swimming','Cricket']
wiki_lst=[]
title=[]
for article in articles:
  print("loading content: ",article)
  wiki_lst.append(wikipedia.page(article).content)
  title.append(article)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words={'english'})
X = vectorizer.fit_transform(wiki_lst)
```

```python
print(vectorizer.get_feature_names())
```

```python
print(X.shape)
```

```python
print(X)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans as kmeans
distances = []
K = [4,6,8]
for k in K:
  kmn = kmeans(n_clusters=k, max_iter=128)
  kmn = kmn.fit(X)
  distances.append(kmn.inertia_)
  print("For K=",k,"J_Clust is",round(kmn.inertia_,4),"\n")
#K=12 isn't possible because we have 9 samples
```

```python
for k in K:
  clus = kmeans(n_clusters=k, max_iter=100)
  clus.fit(X)
  labels=clus.labels_
  wiki=pd.DataFrame(list(zip(title,labels)),columns=['title','cluster'])
  print("document-cluster association for k=",k,"\n",wiki,"\n\n")
```

In this case as we have just 9 documents of different types and among them we can observe Linear Algebra, Data Sciene, Artificial Intelligence are related, European Central bank, Financial Technology, International Monetory Fund are related, Basketball, Swimming and cricket are related, 4 clusters seems to be ideal. So, k=4 should be the best at generalised clustering here
