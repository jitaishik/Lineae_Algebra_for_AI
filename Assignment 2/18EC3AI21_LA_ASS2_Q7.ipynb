Generate a 30 × 10 random matrix A and a 30 × 1 random vector b.
Check that the matrix is full column rank! Run the algorithm for 100
steps. Verify numerically that the algorithm converges to $ \hat{x} $ .

```python
A = np.random.rand(30,10)
b = np.random.rand(30,1)
rank = np.linalg.matrix_rank(A)
print(rank)
```

Since the column rank of the matrix is 10, therefore it is full column rank. 

Next, we calculate the value using direct method and then check if it matches with the numerical solution. 

```python
pseudo_inv = np.dot((np.linalg.inv(np.dot(np.transpose(A), A))), np.transpose(A))
x_direct = np.dot(pseudo_inv,b)
print(x_direct)
```

Next, we can use the iterative LS to calculate the answer: <br> 
<center>
$$ x^{(k+1)} = x^{(k)} - \frac{A^T}{|| A||^2 } (Ax^{(k)} - b)  $$
</center>

```python
A_norm = np.linalg.norm(A, 2)
```

```python
x_init = np.zeros((10,1))
for iter in range(100) :
    x_next = x_init -  (np.dot(np.transpose(A), ((np.dot(A , x_init) - b) ) )/(A_norm ** 2) )
    x_init = x_next  
    dist = np.linalg.norm(x_direct - x_init )
    print(dist)
```

Here, we can see that the distance between the analytical solution and the numerical solution is converging towards 0.

We run it for 1000 iterations next and observe it almost converges to 0

```python
x_init = np.zeros((10,1))
for iter in range(1000) :
    x_next = x_init -  (np.dot(np.transpose(A), ((np.dot(A , x_init) - b) ) )/(A_norm ** 2) )
    x_init = x_next  
    dist = np.linalg.norm(x_direct - x_init )
    print(dist)
```
